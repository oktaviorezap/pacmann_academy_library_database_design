```python
import pandas as pd
import random
from datetime import datetime, timedelta
```

```python
# Function to generate a random date within a range
def random_date(start_date, end_date):
    return start_date + timedelta(days=random.randint(0, (end_date - start_date).days))

# Set the start and end date for the range (e.g., between 5 to 10 years ago)
start_date = datetime(2013, 1, 1).date()  # Start date: January 1, 2013
end_date = datetime(2018, 12, 31).date()  # End date: December 31, 2018

# Generate dummy data for the Library_Managers table
library_managers_data = {
    'manager_id': range(1, 4),
    'first_name': [f'ManagerFirstName{i}' for i in range(1, 4)],
    'last_name': [f'ManagerLastName{i}' for i in range(1, 4)],
    'phone_number': [f'0274-1234{i}' for i in range(1, 4)],
    'email': [f'manager{i}@library.com' for i in range(1, 4)],
    'hire_date': [random_date(start_date, end_date) for _ in range(3)]  # Random hire_date
}

# Create the DataFrame
library_managers_df = pd.DataFrame(library_managers_data)

# Display the DataFrame
library_managers_df
```

```python
# Generate dummy data for the Libraries table
libraries_data = {
    'library_id': range(1, 4),
    'library_name': ['Perpustakaan Pintar', 'Perpustakaan Cerdas', 'Perpustakaan Logika'],
    'library_address': ['Jl. Beo No 230, Bantul, Yogyakarta', 'Jl. Markisa No 240, Bantul, Yogyakarta', 'Jl. Apel No 235, Sleman, Yogyakarta'],
    'phone_number': ['0274-88032', '0274-90902', '0274-98783'],
    'manager_id': [1, 2, 3],
    'library_email': [f'{name.lower().replace(" ", "")}@library.com' for name in ['Perpustakaan Pintar', 'Perpustakaan Cerdas', 'Perpustakaan Logika']],
    'library_website': [f'www.{name.lower().replace(" ", "")}.com' for name in ['Perpustakaan Pintar', 'Perpustakaan Cerdas', 'Perpustakaan Logika']],
    'membership_fee': [random.choice(range(10, 21)) * 1000 for _ in range(3)],  # Rounded thousands
    'library_established_date': [
        (datetime.now() - timedelta(days=random.randint(365*10, 365*50))).strftime('%Y-%m-%d') 
        for _ in range(3)
    ]
}

# Create the DataFrame
libraries_df = pd.DataFrame(libraries_data)

# Display the DataFrame
libraries_df
```

```python
# Generate dummy data for the Library_Book_List table
library_book_list_data = {
    'library_book_list_id': range(1, 31),
    'library_id': random.choices([1, 2, 3], k=30),
    'book_id': random.choices(range(1, 11), k=30),
    'library_book_quantity': random.choices(range(20, 40), k=30)
}
library_book_list_df = pd.DataFrame(library_book_list_data)

# Group by book_id and library_id to show quantities for each library and book combination
book_totals = library_book_list_df.groupby(['library_book_list_id','library_id', 'book_id'])['library_book_quantity'].sum().reset_index()
book_totals
```

```python
# Create Book Groups to calculate total number of books in each libraries
book_groups = book_totals.groupby('book_id')['library_book_quantity'].sum().reset_index()
book_groups
```

```python
# Generate dummy data for the Books table
# Function to generate a random date within a range
def random_date(start_date, end_date):
    return start_date + timedelta(days=random.randint(0, (end_date - start_date).days))

# Set the start and end date for the range (e.g., between 1 to 20 years ago)
start_date = datetime(2003, 1, 1).date()  # Start date: January 1, 2003
end_date = datetime(2023, 1, 1).date()  # End date: January 1, 2023

books_data = {
    'book_id': range(1, 11),
    'title': [f'Book Title {i}' for i in range(1, 11)],
    'author': [f'Author {i}' for i in range(1, 11)],
    'category': random.choices(
        ['Self-Improvement', 'Biography', 'Fantasy', 'Romance', 'Science Fiction', 'History', 'Mystery', 'Non-Fiction', 'Children', 'Horror'], k=10
    ),
    'published_date': [random_date(start_date, end_date) for _ in range(10)],  # Random published_date,
    'isbn': [f'ISBN{i:013}' for i in range(1, 11)]
}
books_df = pd.DataFrame(books_data)

# Merge the total available quantities into the books data
books_df = books_df.merge(book_groups, on='book_id', how='left')
books_df['library_book_quantity'].fillna(0, inplace=True)
books_df.rename(columns={'library_book_quantity': 'available_quantity'}, inplace=True)
books_df['available_quantity'] = books_df['available_quantity'].astype(int)

# Display the DataFrame
books_df = books_df[['book_id', 'title', 'author', 'category','available_quantity','published_date','isbn']]
books_df
```

```python
# Define the date range for registration_date (before 1 January 2023)
start_date = datetime(2021, 1, 1)
end_date = datetime(2022, 12, 31)

# Generate a list of random dates within the specified range
date_range = pd.date_range(start=start_date, end=end_date).to_list()

# Generate dummy data for the User_Registration table
users_data = {
    'user_id': range(1, 21),
    'first_name': [f'UserFirstName{i}' for i in range(1, 21)],
    'last_name': [f'UserLastName{i}' for i in range(1, 21)],
    'email': [f'user{i}@example.com' for i in range(1, 21)],
    'registration_date': random.sample(date_range, k=20)  # Sample random dates from the date_range
}

# Create the DataFrame
users_df = pd.DataFrame(users_data)

# Display the DataFrame
users_df
```

```python
# Generate a list of random dates within a certain period (e.g., last year)
start_date = datetime(2023, 1, 1)
end_date = datetime(2024, 1, 1)
date_range = pd.date_range(start=start_date, end=end_date).to_list()

# Generate dummy data for the Books_Borrow table
books_borrow_data = {
    'book_borrow_id': range(1, 21),
    'user_id': random.choices(range(1, 21), k=20),
    'library_book_list_id': random.choices(range(1, 21), k=20),
    'borrow_date': random.sample(date_range, k=20),  # Sample random dates from the date_range
    'borrow_book_quantity': random.choices(range(1, 3), k=20),
    'books_returned': [True if random.random() > 0.5 else False for _ in range(20)]
}

# Create the DataFrame
books_borrow_df = pd.DataFrame(books_borrow_data)

# Sort the borrow_date column in ascending order
books_borrow_df['borrow_date'] = books_borrow_df['borrow_date'].sort_values().values

# Set the due_date to 14 days after the borrow_date
books_borrow_df['due_date'] = books_borrow_df['borrow_date'] + timedelta(days=14)

# Set the return_date randomly or as None
books_borrow_df['return_date'] = [
    None if random.random() > 0.5 else row['borrow_date'] + timedelta(days=random.randint(1, 14))
    for _, row in books_borrow_df.iterrows()
]

# Display the DataFrame
books_borrow_df
```

```python
# Set a fixed end date for the random generation
end_date = datetime(2024, 1, 1)

# Generate random hold_date within the last 14 days, but order them from earliest to latest
book_holds_data = {
    'book_hold_id': range(1, 21),
    'user_id': random.choices(range(1, 21), k=20),
    'library_book_list_id': random.choices(range(1, 21), k=20),
    'book_hold_quantity': random.choices(range(1, 3), k=20),
    'hold_date': sorted([end_date - timedelta(days=random.randint(1, 14)) for _ in range(20)])  # Sort the random dates
}

# Create the DataFrame
book_holds_df = pd.DataFrame(book_holds_data)

# Set expiry_date to 7 days after the hold_date
book_holds_df['expiry_date'] = book_holds_df['hold_date'] + timedelta(days=7)

# Set is_active randomly (with 70% chance being True)
book_holds_df['is_active'] = [True if random.random() > 0.3 else False for _ in range(20)]

# Display the DataFrame
book_holds_df
```
